```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Plot_2images(imagen01,imagenResultante,imagen02):
  plt.subplot(131)
  plt.imshow(cv2.cvtColor(imagen01, cv2.COLOR_BGR2RGB))
  plt.title('Imagen 1')
  plt.xticks([])
  plt.yticks([])

  plt.subplot(132)
  plt.imshow(cv2.cvtColor(imagenResultante, cv2.COLOR_BGR2RGB))
  plt.title('Imagen Resultante')
  plt.xticks([])
  plt.yticks([])

  plt.subplot(133)
  plt.imshow(cv2.cvtColor(imagen02, cv2.COLOR_BGR2RGB))
  plt.title('Imagen 2')
  plt.xticks([])
  plt.yticks([])

  plt.show()
```

```python
def Plot_1image(imagen01,imagenResultante):
  plt.subplot(131)
  plt.imshow(cv2.cvtColor(imagen01, cv2.COLOR_BGR2RGB))
  plt.title('Imagen 1')
  plt.xticks([])
  plt.yticks([])

  plt.subplot(132)
  plt.imshow(cv2.cvtColor(imagenResultante, cv2.COLOR_BGR2RGB),vmax=255)
  plt.title('Imagen Resultante')
  plt.xticks([])
  plt.yticks([])

  plt.show()
```

```python
def Plot_3hist(hist01,hist02,hist03):
  plt.subplot(131)
  plt.bar(range(len(hist01)), hist01)
  
  plt.subplot(132)
  plt.bar(range(len(hist02)), hist02)

  plt.subplot(133)
  plt.bar(range(len(hist03)), hist03)

  plt.show()
```

```python
def Plot_2hist(hist01,hist02):
  plt.subplot(131)
  plt.bar(range(len(hist01)), hist01)
  
  plt.subplot(132)
  plt.bar(range(len(hist02)), hist02)

  plt.show()
```

```python
'''ECUALIZACION'''

def ecualizacion(img):
  ima_gris=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  imag_ecu = cv2.equalizeHist(ima_gris)
  return imag_ecu
```

```python
'''HISTOGRAMA'''
def histograma(img):
  # Normalización del histograma por su tamaño
  [M, N] = img.shape[0:2]

  # Características del histograma: images, channels, mask, histSize, ranges, hist=..., accumulate=...)
  hist = cv2.calcHist([img], [0], None, [50], [0,255]).flatten()/(M*N)
  #fig=plt.figure()
  return hist
```
